```python
%matplotlib notebook

import numpy as np
import qubitvis as qv
```

```python
# Initialize the qubit
qubit = qv.Qubit()
```

```python
# Plot the qubit
qubit.show()
```

```python
# Set the magnetic field acting on the qubit
qubit.magnetic_field([0,1,0])
```

```python
# Get the Hamiltonian
qubit.hamiltonian()
```

```python
# Show an animation of the qubit precessing
qubit.animate()
```

```python
# Show the entire path the qubit traces
qubit.path()
```

# Dynamical Decoupling Sequences

```python
sequence = qv.Sequence('XI', time_interval=0.5)
```

```python
# Set the initial state to the |+> state
qubit.set_state([[1/np.sqrt(2)], [1/np.sqrt(2)]])
# Set the qubit's pulse sequence
qubit.sequence(sequence)

qubit.show()
```

```python
qubit.animate(dt=0.01)
```

### XY4

```python
qubit.sequence(qv.xy4(0.1))
qubit.magnetic_field([1, 1, 1])
```

```python
# Draw entire path
qubit.path(steps=2000, dt=0.01)
```

```python
# Draw path in steps of iterations of the entire sequence
# This way we can visualize how the state precesses after each iteration of the sequence
qubit.path(dt=0.4)
```
